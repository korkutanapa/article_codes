```python
dpi=['1PWTemperature', '2MWPeriod', '2MWTop_Spray_Circulation_Period',
       '8DSClosed_Cycle_Period', '10DYFan', '10DYFan_Flap',
       '10DYExtra_Heater_Offset']
x_dpi=df[dpi]
x_dpi
```

```python
wc=['1PWClosed_Cycle_Period', '1PWWaterInlet', '1PWRPM', '1PWTemperature',
       '2MWClosed_Cycle_Period', '2MWWaterInlet', '2MWRPM', '4CRPeriod',
       '4CRWaterInlet', '4CRTop_Spray_Circulation_Period',
       '7RSClosed_Cycle_Period', '7RSPeriod', '7RSRPM',
       '7RSLower_Spray_Circulation_Period', '8DSClosed_Cycle_Period',
       '10DYPeriod', '10DYExtra_Heater_Offset']
x_wc=df[wc]
x_wc
```

```python
model_gbr = GradientBoostingRegressor(learning_rate = 0.1, n_estimators = 100, max_depth = 4)
```

```python
model_knn =KNeighborsRegressor(n_neighbors=3)
```

```python
model_lr=LinearRegression()
```

```python
X = x_cpi
y = y_CPI
model_gbr.fit(X, y)
filename = 'cpi.sav'
joblib.dump(model_gbr, filename)

cv = RepeatedKFold(n_splits=5, n_repeats=3, random_state=79)
scores1 = cross_val_score(model_gbr,X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1, error_score='raise')
scores3 = cross_val_score(model_gbr,X, y, scoring='r2', cv=cv, n_jobs=-1, error_score='raise')
print(' %.3f %.0f' % (-mean(scores1),  100*mean(scores3)))
```
